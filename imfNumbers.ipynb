```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
nBins=128
ms=10**np.linspace(-2,2,nBins)
mSwitch=np.argmin(np.abs(ms-1))
epsLow=0.158*np.exp(-np.power(np.log10(ms[:mSwitch+1]) - np.log10(0.08),2) / (2*(0.69**2)))/(ms[:mSwitch+1]*np.log(10))
epsHigh=epsLow[-1]*np.power(ms[mSwitch+1:],-2.3)
eps=np.hstack((epsLow,epsHigh))
print(eps.shape)
```

```python
eps=np.hstack((epsLow,epsHigh))
print(eps.shape)
```

```python
eps=np.zeros(nBins-2)
for i in range(nBins-2):
    eps[i]=epsLow[i]*((ms[i]+ms[i+2])*np.log(10))/2 #in theory in Msun pc-3
```

```python
plt.semilogx(ms,eps)
plt.show()
```

```python
plt.loglog(ms,eps)
plt.show()
```

```python
data=np.vstack((ms,eps)).T
print(data.shape)
```

```python
footerText='/Chabrier IMF/#3D8EDB/--/'
np.savetxt('data/chabrierImf.txt',data,fmt='%1.3e \t',header='M (M_s) \t dN/dMdV (M_s^-1 pc^-3)',footer=footerText)
```
